```python
N = 512
D = 256
a = 0.125

g = 1
v, Q = 1, 1
D0, tol, method = D, 1e-6, '12site'
#
ms = [0 * g]
Nas =  [(N, a)]

data = {}

for m in ms:
    for N, a in Nas:
        dt = 1 / 16
        folder = folder_evol(g, m, a, N, v, Q, D0, dt, D, tol, method)
        data[m, N, a, D, dt] = np.load(folder / f"results.npy", allow_pickle=True).item()
```
